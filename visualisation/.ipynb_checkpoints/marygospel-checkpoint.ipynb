```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv('housing.csv')
dataset.head()
```

```python
#Histogram to show price distribution
sns.histplot(data = dataset, x = 'Price');
```

```python
# Scatterplot to show the relationship of prive and income
sns.scatterplot(data=dataset, x='Avg. Area Income', y='Price')
plt.title('Avg. Area Income vs Price')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')   
```
