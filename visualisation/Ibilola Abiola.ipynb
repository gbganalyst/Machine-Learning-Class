```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
diabetes = pd.read_csv('data_cleaningML/diabetes.csv')
```

```python
diabetes.head()
```

```python
# scatter plot for blood pressure vs AGE
plt.Figure(figsize=(10,6))
sns.scatterplot(data=diabetes, x='BloodPressure', y = 'Age')
plt.title('Scatter Plot of Blood Pressure vs Age')
plt.xlabel('Blood Pressure')
plt.ylabel('Age')
plt.show()
```
