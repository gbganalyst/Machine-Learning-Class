```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

Housing = pd.read_csv('housing.csv')

```

```python
Housing.head()
```

```python
Housing = Housing.rename(columns={'Avg. Area Income': 'AvgAreaIncome'})
Housing.head() 
```

```python
Housing.columns
```

```python
sns.scatterplot(x="AvgAreaIncome", y="Price", hue="Price", palette="turbo", data=Housing, alpha= 0.5)
plt.figure(figsize=(10, 6))
plt.title("AverageAreaIncome vs Price")
plt.xlabel("Average_Area_Income")
plt.ylabel("Price")
plt.show()
```

```python
Housing[['AvgAreaIncome', 'Price']].corr() 
```

```python
sns.histplot(Housing['Price'], bins=30, kde=True, color="blue",alpha=0.5)
plt.title("Frequency of House Prices")

plt.show()
```
