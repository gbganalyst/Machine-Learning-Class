```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# Load dataset

diabetes = pd.read_csv('../data_cleaningML/diabetes.csv')
```

```python
diabetes.head()
```

```python
# plot scatter plot for blood pressure vs bmi

sns.scatterplot(data=diabetes, x='BloodPressure', y='BMI')
plt.title('Blood Pressure vs BMI')
plt.xlabel('Blood Pressure')
plt.ylabel('BMI')   
plt.show()
```

```python
diabetes[['BloodPressure', 'BMI']].corr()
```

```python
# Distribution of Blood pressor

sns.histplot(data = diabetes, x = 'BloodPressure');
plt.show()
```

```python
#Distribution of Blood pressure vs BMI using line chart

sns.lineplot(data=diabetes, x='BloodPressure', y='BMI')
plt.title('Blood Pressure vs BMI')
plt.xlabel('Blood Pressure')
plt.ylabel('BMI')   
plt.show()
```
