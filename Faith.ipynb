```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# hdt stands for heart data
hdt = pd.read_csv('data_cleaningML\heart.csv')
```

```python
# This is to have an idea of the contents of the dataset
hdt.head()
```

```python
# This is to figure out the features I have to work with
hdt.columns
```

```python
# Scatter Plot

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

plt.scatter(hdt['age'], hdt['thalach'])

plt.title('Age vs Maximum Heart Rate')
plt.xlabel('Age (years)')
plt.ylabel('Maximum Heart Rate Achieved')

plt.show()
```

```python
# Boxplot

plt.figure(figsize=(8, 5))

hdt.boxplot(column='chol', by='target')

plt.title('Cholesterol Levels by Heart Disease Status')
plt.suptitle('')  # removes automatic pandas title
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol')

plt.show()
```
