**Importing the python libaries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Importing the data to be plotted**

```python
data = pd.read_csv('diabetes.csv')
```

```python
data.head()
```

**Plotting Age vs Glucose Level**

```python
plt.plot(data['Age'], data['Glucose'],'^', color='blue')
plt.xlabel('Age')
plt.ylabel('Glucose Level')
plt.title('Age vs Glucose Level')
plt.show()
```
